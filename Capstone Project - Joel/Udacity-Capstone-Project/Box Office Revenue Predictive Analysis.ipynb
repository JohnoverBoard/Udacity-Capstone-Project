The data we have is in the format of `mm/dd/yy` but for processing, we need to convert it to atleast `mm/dd/yyyy` to avoid year ambiguity.

```python
df_movie['release_date'].head(10)
```

**Inference**: The last two digits is ambiguous as it may be a year from the 20th century or 21st century. Let us convert this to python date format to make later stage processing easier.

```python
def date(x):
    x=str(x)
    year=x.split('/')[2]
    if int(year)<19:
        return x[:-2]+'20'+year
    else:
        return x[:-2]+'19'+year

#Clarify the year of release   
df_movie['release_date']=df_movie['release_date'].fillna('1/1/90').apply(lambda x: date(x))
```

### 4. Feature Engineering

Performing the necessaary pre-processing actions on the dataset.

```python
def  prepare_data(df):
    
    #Taking suitable ratios as concluded to be useful from the above graphs
    df['_budget_runtime_ratio'] = (df['budget']/df['runtime']).replace([np.inf,-np.inf,np.nan],0)
    df['_budget_popularity_ratio'] = df['budget']/df['popularity']
    df['_budget_year_ratio'] = df['budget'].fillna(0)/(df['release_year']*df['release_year'])
    df['_releaseYear_popularity_ratio'] = df['release_year']/df['popularity']
    df['_releaseYear_popularity_ratio2'] = df['popularity']/df['release_year']
    df['budget']=np.log1p(df['budget'])
    
    #Create a new column and set its values to those of belongs_to_collection if applicable
    df['collection_name']=df['belongs_to_collection'].apply(lambda x: x[0]['name'] if x != {} else 0)
    
    #Create a new column and set its values to 0 unless a homepage exists
    df['has_homepage']=0
    df.loc[(pd.isnull(df['homepage'])),'has_homepage']=1
    
    #Encoding the collection_name column
    le=LabelEncoder()
    le.fit(list(df['collection_name'].fillna('')))
    df['collection_name']=le.transform(df['collection_name'].fillna('').astype(str))
    
    #Encoding the original_language column
    le=LabelEncoder()
    le.fit(list(df['original_language'].fillna('')))
    df['original_language']=le.transform(df['original_language'].fillna('').astype(str))
    
    #Set _num_Keywords column as the value of number of keywords
    df['_num_Keywords'] = df['Keywords'].apply(lambda x: len(x) if x != {} else 0)
    
    #Set _num_cast column as the value of number of actors
    df['_num_cast'] = df['cast'].apply(lambda x: len(x) if x != {} else 0)
    
    #Set isbelongto_coll column as 0 unless that row does belong to a collection, in which case set a 1.
    df['isbelongto_coll']=0
    df.loc[pd.isna(df['belongs_to_collection']),'isbelongto_coll']=1
    
    #Set isTaglineNA column as 0 unless that row does have a tagline, in which case set a 1.
    df['isTaglineNA'] = 0
    df.loc[df['tagline'] == 0 ,"isTaglineNA"] = 1 

    #Set isOriginalLanguageEng column as 0 unless that movie is indeed in English, in which case set a 1.
    df['isOriginalLanguageEng'] = 0 
    df.loc[ df['original_language'].astype(str) == "en" ,"isOriginalLanguageEng"] = 1
    
    #Set isbelongto_coll column as 1 unless that movie is not released, in which case set a 0.
    df['ismovie_released']=1
    df.loc[(df['status']!='Released'),'ismovie_released']=0
    
    #Set the number of spoken languages under the new no_spoken_languages column
    df['no_spoken_languages']=df['spoken_languages'].apply(lambda x: len(x))
    
    #Set the original title letter count under the new original_title_letter_count column
    df['original_title_letter_count'] = df['original_title'].str.len() 
    
    #Set the original title word count under the new original_title_word_count column
    df['original_title_word_count'] = df['original_title'].str.split().str.len() 

    #Set the title word count under the new title_word_count column
    df['title_word_count'] = df['title'].str.split().str.len()
    
    #Set the overview word count under the new overview_word_count column
    df['overview_word_count'] = df['overview'].str.split().str.len()
    
    #Set the tagline word count under the new tagline_word_count column
    df['tagline_word_count'] = df['tagline'].str.split().str.len()
    
    #Set the id of collection under the new collection_id column if applicable
    df['collection_id'] = df['belongs_to_collection'].apply(lambda x : np.nan if len(x)==0 else x[0]['id'])
    
    #Set the number of production countries under the new production_countries_count column if applicable
    df['production_countries_count'] = df['production_countries'].apply(lambda x : len(x))
    
    #Set the number of production companies under the new production_companies_count column if applicable
    df['production_companies_count'] = df['production_companies'].apply(lambda x : len(x))
    
    #Set the number of actors under the new cast_count column if applicable
    df['cast_count'] = df['cast'].apply(lambda x : len(x))
    
    #Set the number of crew members under the new crew_count column if applicable
    df['crew_count'] = df['crew'].apply(lambda x : len(x))

    #Set the gender of each crew member
    df['genders_0_crew'] = df['crew'].apply(lambda x: sum([1 for i in x if i['gender'] == 0]))
    df['genders_1_crew'] = df['crew'].apply(lambda x: sum([1 for i in x if i['gender'] == 1]))
    df['genders_2_crew'] = df['crew'].apply(lambda x: sum([1 for i in x if i['gender'] == 2]))

    #Create a dedicated column for each of the values
    for col in  ['genres', 'production_countries', 'spoken_languages', 'production_companies'] :
        df[col] = df[col].map(lambda x: sorted(list(set([n if n in movie_dict[col] else col+'_etc' for n in [d['name'] for d in x]])))).map(lambda x: ','.join(map(str, x)))
        temp = df[col].str.get_dummies(sep=',')
        df = pd.concat([df, temp], axis=1, sort=False)
    
    #Remove the genres_etc column
    df.drop(['genres_etc'], axis = 1, inplace = True)
    
    #Normalize the columns present in the dataset
    cols_to_normalize=['runtime','popularity','budget','_budget_runtime_ratio','_budget_year_ratio','_budget_popularity_ratio','_releaseYear_popularity_ratio',
    '_releaseYear_popularity_ratio2','_num_Keywords','_num_cast','no_spoken_languages','original_title_letter_count','original_title_word_count',
    'title_word_count','overview_word_count','tagline_word_count','production_countries_count','production_companies_count','cast_count','crew_count',
    'genders_0_crew','genders_1_crew','genders_2_crew']
    for col in cols_to_normalize:
        print(col)
        x_array=[]
        x_array=np.array(df[col].fillna(0))
        X_norm=normalize([x_array])[0]
        df[col]=X_norm
    
    #Remove the previously used columns as it is now redundant due to better informational columns
    df = df.drop(['belongs_to_collection','genres','homepage','imdb_id','overview','id'
    ,'poster_path','production_companies','production_countries','release_date','spoken_languages'
    ,'status','title','Keywords','cast','crew','original_language','original_title','tagline', 'collection_id'
    ],axis=1)
    
    #Fill empty positions, if any, with 0.0
    df.fillna(value=0.0, inplace = True) 

    return df

    
```

To deal with JSON Files:

```python
def get_json(df):
    global dict_columns
    result=dict()
    for col in dict_columns:
        d=dict()
        rows=df[col].values
        for row in rows:
            if row is None:
                continue
            for i in row:
                if i['name'] not in d:
                    d[i['name']]=0
                else:
                    d[i['name']]+=1
            result[col]=d
    return result

movie_dict=get_json(df_movie)
```
